```python
# sin함수에 오차가 가장 적은 직선 그리기

x = np.arange(-3, 3, 0.01)  # 문제집
y = np.sin(x) #

a = 0
b = 0
c = 0
d = 0
lr = 0.000005  # lr은 학습률(learning rate)로, 각 파라미터를 업데이트할 때 얼마나 크게 업데이트할지를 조절합니다.

```

```python
for n in range(3000):
  y_hat = a * x**3 + b * x**2 + c * x + d  # 3차 함수 모델
  error = y_hat - y #예측값 y_hat과 실제값 y 사이의 오차를 나타냅니다.

    # 편미분 값을 계산합니다. 이 편미분 값들은 경사 하강법의 업데이트 규칙에서 사용됩니다.
  a_diff = sum(2 * error * x**3)
  b_diff = sum(2 * error * x**2)
  c_diff = sum(2 * error * x)
  d_diff = sum(2 * error * 1)

  a = a - a_diff * lr
  b = b - b_diff * lr
  c = c - c_diff * lr
  d = d - d_diff * lr

  print(f"{n+1}회 학습 // {a_diff:.2f}, {b_diff:.2f}, {c_diff:.2f}, {d_diff:.2f}")
```

```python
# 그래프 그리기
plt.plot(x, y, label='3cubic function')
plt.plot(x, a * x**3 + b * x**2 + c * x + d, label=f'third degree polynomial', linestyle='--', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic functions and cubic polynomial graphs')
plt.grid(True)
plt.show()
```

```python
from  tensorflow import keras
from sklearn.model_selection import train_test_split
import pandas as pd


df = pd.read_csv("bike.csv")
df
```

```python
del df['registered']
del df['datetime']
df
```

```python
data = df.iloc[:, :-1].to_numpy()
target = df['count'].to_numpy()
```

```python
# 데이터 표준화
# 데이터 split
from sklearn.preprocessing import StandardScaler
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)

ss = StandardScaler()
train_scaled = ss.fit_transform(train_input)
test_scaled = ss.transform(test_input)

model = keras.Sequential() # 도화지 한장 만드는, 인공신경망을 만들기 위한...
model.add(keras.layers.Dense(1, input_shape = (9,))) # input_shape은 입력층임, 한개의 입력층에서 하나의 출력층으로 간다는 뜻)
# 총 컴퓨터가 학습해야 되는 파라미터는 10개!! 9개의 컬럼과 1개의 절편. model.summary() 코드로 확인가능!!!
model.compile(loss= 'mse', optimizer = 'adam', metrics = 'mae') # adam은 학습률을 직접 자동으로 조절해준다!! # metrics = 'mae' 절대 오차를 보여준다.
# 컴퓨터한테 mse 방법으로 계산해달라고 하고, 학습률을 자동으로 맞춰주게 amda을 사용하였고, 거기서 내가 보고 싶은 데이터인 절대오차를 보여달라고 metrics = 'mae'를 사용한거

# MSE 평균제곱오차

es = keras.callbacks.EarlyStopping(patience=5, restore_best_weights = True) # patience=5 : 컴퓨터가 참는 횟수(과대적합이 이어지는 수 현재는 5번연속 과대적합일때, 5번연속 모델이 좋아지지 않았을때, 주기!),
# restore_best_weights = True, 과대적합이(모델이 5번연속 모델이 좋아지지 않앗을때) 5번 연속 일어났을때, 그것을 제외하고(이전의) 모델을 보여줘!라는 뜻

hist = model.fit(train_scaled, train_target, epochs = 100, validation_data = (test_scaled, test_target), callbacks=[es], batch_size = 100) # epochs 학습 횟수!
# validation_data = (test_scaled, test_target) # 쪽지시험, 훈련을 한번 할때마다 쪽지 시험, 훈련데이터는 떨어지는, 쪽지시험은 올라가는 과대적합을 확인 할 수 있음!!
# val_loss인 쪽지시험의 결과를 확인 해야된다!!! - > 과대적합을 확인했을 경우 -> 얼리스탑핑 적용, 이렇게 하여, 최적의 모델을 확인 할 수 있음??!
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.show()
```

```python
# 출력층에 시그모이드 함수가 붙어있고, 여기를 통해서 나온 s1,s2,s3 데이터를 확인하면, 0.8, 0.7, 0.6 이렇게 나와있는데,
# 이것을 분모에 다 더하고, 확인하고자 하는 것을 데이터를 분자에 올리면, s1,s2,s3를 확률로 다 더하면 1인 것을 확인 할 수 있으며, 이것을 소프트 맥스 방법이라고 하며,
# 이 상태에서 다시 모델을 돌려 계속 학습을 시키는 것을 , 크로스엔트로프라고 할 수 있다!!
```

```python
df = pd. read_csv('pima_indians.csv') # class가 0이면 노당요, 1이면 당요 , 정답지 데이터이다!

data = df.iloc[:,:-1].to_numpy()
target = df['Class'].to_numpy()

# 데이터 표준화
# 데이터 split
from sklearn.preprocessing import StandardScaler
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)

ss = StandardScaler()
train_scaled = ss.fit_transform(train_input)
test_scaled = ss.transform(test_input)


# 모델 만들어주기
model = keras.Sequential()
model.add(keras.layers.Dense(32, activation = 'relu', input_shape =(8,))) # softmax - > 확률로 보겠다!
# 드랍아웃, 32개의 은닉충 중 랜덤으로 40%를 끊게한다.  - > 컴퓨터의 학습을 방해 -> 과대적합을 방해한다는 이유로 사용 -> 억누른다. - >드랍아웃을 40프로 넣었다는 것은 거의 최대치로 넣어것.  - >
# - > 이 상태로 그림을 그려서 확인 후에, 은닉층을 더 할지 뺄지 결정해야 한다...이 부분은 많은 경험이 필요할듯..
model.add(keras.layers.Dropout(0,4))
model.add(keras.layers.Dense(16, activation = 'relu'))
# 드랍아웃, 32개의 은닉충 중 랜덤으로 40%를 끊게한다.
model.add(keras.layers.Dropout(0,4))
model.add(keras.layers.Dense(2, activation = 'softmax'))
# 정답지가 0,1 처럼 원핫인코딩이 되어 있으면 아래 코드 에러, 이유로는 loss= sparse부분이 원핫인코딩을 같이해서 보내주겠다라는 뜻!
# 그래서 원핫인코딩이 안 되어 있는 target값에만 sparse_를 같이 붙여준다.
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics = 'accuracy')

es = keras.callbacks.EarlyStopping(patience=5, restore_best_weights = True) # patience=5 : 컴퓨터가 참는 횟수(과대적합이 이어지는 수 현재는 5번연속 과대적합일때, 5번연속 모델이 좋아지지 않았을때, 주기!),

hist = model.fit(train_scaled, train_target, epochs = 500, validation_data = (test_scaled, test_target)) # epochs 학습 횟수!
# validation_data = (test_scaled, test_target) # 쪽지시험, 훈련을 한번 할때마다 쪽지 시험, 훈련데이터는 떨어지는, 쪽지시험은 올라가는 과대적합을 확인 할 수 있음!!
# val_loss인 쪽지시험의 결과를 확인 해야된다!!! - > 과대적합을 확인했을 경우 -> 얼리스탑핑 적용, 이렇게 하여, 최적의 모델을 확인 할 수 있음??!
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.show()

# 과대적합을 볼 수 있다!!!
# 간단한 데이터에 강력한 모델을 집어넣어서!!
# 오차가 줄어들다, 과대적합을 확인 할 수 잇다!
# 과대적합 이유
# 1. 모델이 복잡 하거나
# 2. activation이 sigmoid여서?, 분류 층에는 마지막은 무조건 sigmoid? 확률로 보여주기 때문에
# 3. 시그모이드 미분을 하면 70%가 날아간다!! - > 은닉층(sigmoid가 적용된)을 지날 때 마다, 데이터가 손실을 한다.
# 4. 3.문제를 없에기 위해서, 렐루라는 것을 적용시킨다(은닉층!에서) - > 이러한 문제를 기울기 소실 문제라고 한다!!!
```
